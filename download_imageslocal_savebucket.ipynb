```python
from google.cloud import storage
import os
from collections import defaultdict
CONFIG_FILE_NAME = os.getenv("CONFIG_FILE_NAME")

#load data
BUCKET_NAME = 'building_images_storage'

storage_client = storage.Client()
bucket = storage_client.bucket(BUCKET_NAME)

# Create a dictionary to store folders and their files
folder_files = defaultdict(list)

# List all blobs in the bucket
blobs = bucket.list_blobs()

# Organize files by folder
for blob in blobs:
    # Skip if blob is a folder placeholder
    if blob.name.endswith('/'):
        continue
        
    # Get folder path by removing filename
    folder = os.path.dirname(blob.name)
    if folder == '':
        folder = 'root'
        
    # Add filename to appropriate folder list
    filename = os.path.basename(blob.name)
    folder_files[folder].append(filename)

# Convert defaultdict to regular dict
folder_files = dict(folder_files)
folder_files
```

```python
local_folder_path = '/Users/luismdecarvajal/Desktop/dataset_fotos/clasica'
bucket_files = folder_files['fotos_fincas/clasica']
BUCKET_NAME = 'building_images_storage'
storage_client = storage.Client()
bucket = storage_client.bucket(BUCKET_NAME)

# Get list of all files in local folder
local_files = os.listdir(local_folder_path)


    
# Check each local file
for filename in local_files:
    # Skip if not an image file
    if not filename.lower().endswith(('.png', '.jpg', '.jpeg', '.gif')):
        continue
        
    # Check if file exists in bucket
    if filename not in bucket_files:
        print(f"Uploading {filename}...")
        
        # Create blob and upload file
        blob = bucket.blob(filename)
        file_path = os.path.join(local_folder_path, filename)
        blob.upload_from_filename(file_path)
        
        print(f"Successfully uploaded {filename}")
    else:
        print(f"Skipping {filename} - already exists in bucket")
```
